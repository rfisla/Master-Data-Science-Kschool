## Exercise_1: 
Compare models’ predictions and decide which is the best one with different metrics.

```python
pwd
```

```python
import pandas as pd
a = pd.read_excel('exercise_1.xlsx')
a
```

```python
a.columns
```

- Calculamos los valores residuales

```python
a['dif1'] = a['real_value'] - a ['pred_1']
a['dif2'] = a['real_value'] - a ['pred_2']
a['dif3'] = a['real_value'] - a ['pred_3']
```

```python
a
```

### Cálculo BIAS de los 3 modelos

```python
bias1= a.dif1.sum()/len(a)
bias2= a.dif2.sum()/len(a)
bias3= a.dif3.sum()/len(a)
f'El  BIAS del modelo 1 es: {round(bias1,2)} , del modelo 2 es: {round(bias2,2)} ,y del modelo 3 es: {round(bias3,2)}'
```

### Cálculo MAE de los 3 modelos

```python
mae1 = a.dif1.abs().sum() / len(a)
mae2 = a.dif2.abs().sum() / len(a)
mae3 = a.dif3.abs().sum() / len(a)
f'El  MAE del modelo 1 es: {round(mae1,2)} , del modelo 2: {round(mae2,2)} ,y del modelo 3: {round(mae3,2)}'
```

### Cálculo RMSE de los 3 modelos

```python
import math
```

```python
rmse1= math.sqrt((a['dif1']**2).sum() / len(a))
rmse2= math.sqrt((a['dif2']**2).sum() / len(a))
rmse3 = math.sqrt((a['dif3']**2).sum() / len(a))
f'El  RMSE del modelo 1 es {round(rmse1,2)} , del modelo 2 {round(rmse2,2)} ,y del modelo 3 {round(rmse3,2)}'
```

## Ejercicio2:
Build the confusion matrix and calculate main metrics about it. 
a. Which is the best model in general? 
b. Which identify better the 1 class?


```python
import pandas as pd
b = pd.read_excel('exercise_2.xlsx')
b
```

```python
b.columns
```

```python
from sklearn.metrics import confusion_matrix
```

### Matriz de confusión Modelo 1

```python
y_true= b['real_value']
y_pred = b['pred_1']
cm1=confusion_matrix(y_true,y_pred)
cm1
```

```python
TN=cm1[0,0]
FP=cm1[0,1]
FN=cm1[1,0]
TP=cm1[1,1]
```

- Acuraccy, sensitivity & precision Modelo 1

```python
acuraccy1= (TP+TN) / len(b)
sensitivity1= TP/(TP+FN)
precision1= TP/(TP+FP)
f' MODELO 1: Acuraccy: {acuraccy1}, Sensitivity: {round(sensitivity1,2)}, Precision: {round(precision1,2)}'

```

### Matriz de confusión Modelo 2

```python
y_true= b['real_value']
y_pred = b['pred_2']
cm2=confusion_matrix(y_true,y_pred)
cm2
```

```python
TN=cm2[0,0]
FP=cm2[0,1]
FN=cm2[1,0]
TP=cm2[1,1]
```

- Acuraccy, sensitivity & precision Modelo 2

```python
acuraccy2= (TP+TN) / len(b)
sensitivity2= TP/(TP+FN)
precision2= TP/(TP+FP)

f' MODELO 2: Acuraccy: {acuraccy2}, Sensitivity: {round(sensitivity2,2)}, Precision: {round(precision2,2)}'
```
